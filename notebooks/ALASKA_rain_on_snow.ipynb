```python
drive.mount('/content/drive')

def transferir_para_o_drive(nome):
    local_salvamento = "/content/drive/MyDrive/rain_on_snow_NASA/"
    local_colab_arquivo = "./"+nome

    # Mover o arquivo para a pasta desejada no Google Drive
    shutil.move(local_colab_arquivo, local_salvamento)

    return "Arquivo movido com sucesso para o Google Drive."
```

```python
import imageio

file_name = 'ABoVE.Rain-On-Snow.2003_Water_Year_Ah000-001v000-001.001.10212018.tif'
file_path = f'/content/drive/MyDrive/rain_on_snow_NASA/{file_name}'

# Carregar o arquivo de imagem TIFF
image = imageio.imread(file_path)

# Exibir informações sobre a imagem
#print(f"Formato da imagem: {image.format}")
print(f"Tipo de dados da imagem: {image.dtype}")
print(f"Dimensões da imagem: {image.shape}")

import matplotlib.pyplot as plt

# Exibir a imagem usando matplotlib
plt.imshow(image, cmap='gray')  # Use 'gray' colormap para imagens em tons de cinza
plt.axis('off')  # Desativa os eixos
plt.title('Imagem TIFF')  # Define um título para a imagem
plt.show()
```
